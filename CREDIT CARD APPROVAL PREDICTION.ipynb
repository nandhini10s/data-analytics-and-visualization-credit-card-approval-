## CREDIT CARD APPROVAL PREDICTION: DATA ANALYTICS FOR RISK ASSESSMENT AND DECISION MAKING 

```python
#IMPORTING LIBRARIES
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

```python
#IMPORTING THE DATASET
data=pd.read_csv('C:/Users/hp/Desktop/PROJECT/CREDIT CARD/credit_card_approval.csv')
data
```

```python
#FINDING WHETHER THERE ARE ANY NULL VALUES
data.isnull().sum()
```

```python
#DROPPING IRRELEVANT COLUMNS FROM THE DATASET
data=data.drop(['Ethnicity','DriversLicense','ZipCode'],axis=1)
data
```

## DESCRIPTIVE STATISTICS

```python
data[['Age','Debt','YearsEmployed','Income']].describe()
```

## CORRELATION

```python
data[['Debt','Employed','CreditScore','Income','Approved']].corr()
```

## DATA PREPROCESSING

```python
label_encoder=LabelEncoder()
```

```python
data['Industry_encoded']= label_encoder.fit_transform(data['Industry'])
print(data)
```

```python
data['Citizen_encoded']= label_encoder.fit_transform(data['Citizen'])
print(data)
```

```python
data
```

```python
#Cleaning the dataset
data=data.drop(['Industry','Citizen'], axis=1)
data
```

```python
#IMPORTING LIBRARIES
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
x=data[['Gender','Age','Debt','Married','BankCustomer','YearsEmployed','PriorDefault','Employed','Income','Industry_encoded','Citizen_encoded']]
y=data['Approved']
```

```python
# TRAIN TEST SPLIT
#TEST SIZE=0.25
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=42)
```

```python
len(x_train),len(x_test),len(y_train),len(y_test)
```

```python
# RESCALING x_train and x_test TO NORMALIZE THE DATA
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler(feature_range=(0,1))
rescaledx_train = scaler.fit_transform(x_train)
rescaledx_test = scaler.fit_transform(x_test)
```

## LOGISTIC REGRESSION

```python
model=LogisticRegression()
model.fit(rescaledx_train,y_train)
```

```python
#MAKING PREDICTION
y_predict=model.predict(rescaledx_test)
y_predict
```

```python
#Logistic Regression
import statsmodels.api as sm
x1=sm.add_constant(x)
result=sm.Logit(y,x1).fit()
result.summary()
```

## EVALUATION OF MODEL

```python
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
#Accuracy score
accuracy=accuracy_score(y_test, y_predict)
print('Accuracy:',accuracy)
```

```python
#Confusion matrix
confusion=confusion_matrix(y_test, y_predict)
print('Confusion Matrix:')
print(confusion)
```

```python
#Classification report
from sklearn.metrics import classification_report
print(classification_report(y_test,y_predict))
```
